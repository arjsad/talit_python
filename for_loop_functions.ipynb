**Aufgabe 3**

Der Modulo-Operator `%` berechnet den Rest einer Division.

```python
def is_even(n):
    if n % 2 == 0:
        return True
    else:
        return False
```

**Aufgabe 4**


```python
def average(li):
    total = 0
    
    for zahl in li:
        total = total + zahl
        
    if len(li) > 0:
        durchschnitt = total / len(li)
        return durchschnitt
    else:
        return 0
    
average([4, 6, 7, 24, 90, 41])
```
