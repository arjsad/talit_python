Brüche in der Mathe wie $\frac{\pi^2}{6}$ sind besser als Beinbrüche.

```python
import math

n = 1000000000
eulische_zahl = 0

for k in range(n):
    eulische_zahl += 1 / math.factorial(k)

print(eulische_zahl)
```
