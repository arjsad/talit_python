Brüche in der Mathe wie $\frac{\pi^2}{6}$ sind besser als Beinbrüche.

```python
import math

n = 10000
eulische_zahl = 0

for k in range(n):
    eulische_zahl += 1 / math.factorial(k)

print(eulische_zahl)
```

### Fehlerbehandlung mit `try-except` in Python

Das `try-except`-Konstrukt wird verwendet, um Laufzeitfehler (Exceptions) abzufangen, damit das Programm nicht unerwartet abstürzt.

* **`try`-Block:** Hier steht der Code, der potenziell einen Fehler auslösen könnte.
* **`except`-Block:** Hier steht der Code, der ausgeführt wird, wenn im `try`-Block ein bestimmter Fehler auftritt.

**Beispiel:**


## Version 1

```python
try:
    zaehler = float(input("Gib den Nenner der Division ein."))
nenner = float(input("Gib den Zähler der Division ein."))
print("Quotient: ",zaehler/nenner)

except:
    print("Es gab einen Fehler bei der Eingabe oder Berechnung.")
```
