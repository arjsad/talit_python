```python
def local_part(email):
    count = 0
    for element in email:
        if element == '@':
            return email[:count]
        count = count + 1
    

print(local_part('arsadiku@ksr.ch'))

def split_email_address(email):
    count = 0
    for element in email:
        if element == '@':
            return email[:count] , email[count+1:] #alles vor dem @, alles nache dem @, das @ befindet sich auf position 5
        count = count + 1
    

print(split_email_address('arsadiku@ksr.ch'))
```
